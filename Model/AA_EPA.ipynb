```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data/new_data.csv')
data
```

```python
data['region']=['j']*158+['w']*149
data
```

```python
AA=data['C20:4n-6, AA']
EPA=data['C20:5n-3, EPA']
ratio=AA/EPA
region=data['region']
dd={'AA/EPA':ratio, 'region':region,'AA':AA,'EPA':EPA}
df = pd.DataFrame(data=dd)
df
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x="AA", y="AA/EPA", data=df, hue="region")
```

```python
sns.scatterplot(x="EPA", y="AA/EPA", data=df, hue="region")
```

```python
from sklearn.cluster import KMeans
```

```python
aa_epa=np.array((df['AA/EPA']-df['AA/EPA'].mean())/df['AA/EPA'].std())
kmeans = KMeans(n_clusters=2, random_state=0).fit(np.array(df['AA/EPA']).reshape(-1,1))
```

```python
kmeans.labels_
```

```python
j_tn=158
w_tn=149
j_c=0
w_c=0
j_d=0
w_d=0
for i in range (len(kmeans.labels_)):
    if kmeans.labels_[i]==0:
        if df['region'][i]=='j':
            j_c+=1
        else:
            w_d+=1
    else:
        if df['region'][i]=='w':
            w_c+=1
        else:
            j_d+=1

print(max((j_c/j_tn + w_c/w_tn)/2,(j_d/j_tn + w_d/w_tn)/2))
```
