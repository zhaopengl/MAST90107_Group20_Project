```python
data1['result']=['j']*65+['w']*52
data1
```

```python
AA=data1['C20:4n-6, AA']
EPA=data1['C20:5n-3, EPA']
ratio=AA/EPA
result=data1['result']
dd={'AA/EPA':ratio, 'result':result,'AA':AA,'EPA':EPA}
df = pd.DataFrame(data=dd)
df
```

```python
from sklearn.cluster import KMeans
```

```python
aa_epa=np.array((df['AA/EPA']-df['AA/EPA'].mean())/df['AA/EPA'].std())
kmeans = KMeans(n_clusters=2, random_state=0).fit(np.array(df['AA/EPA']).reshape(-1,1))
```

```python
kmeans.labels_
```

```python
j_tn=65
w_tn=52
j_c=0
w_c=0
j_d=0
w_d=0
for i in range (len(kmeans.labels_)):
    if kmeans.labels_[i]==0:
        if df['result'][i]=='j':
            j_c+=1
        else:
            w_d+=1
    else:
        if df['result'][i]=='w':
            w_c+=1
        else:
            j_d+=1

print(j_c/j_tn)
print(w_c/w_tn)
print(j_d/j_tn)
print(w_d/w_tn)

print(max((j_c/j_tn + w_c/w_tn)/2,(j_d/j_tn + w_d/w_tn)/2))


```
