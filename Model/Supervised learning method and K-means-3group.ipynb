```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit, GridSearchCV, cross_val_score
```

```python
data = pd.read_excel('final.xlsx')
X = data.iloc[:,0:23]
data
```

```python
# Logistic Regression
# X_train, X_test, y_train, y_test = train_test_split(X, data.result, test_size=0.2)
# logis = LogisticRegression().fit(X_train, y_train)
# y_test_pre = logis.predict(X_test)
# accuracy_score(y_test, y_test_pre )

logis = LogisticRegression()
print(np.mean(cross_val_score(logis, X, data.result, cv = 10)))
```

```python
# SVM

svm1 = SVC(kernel = "rbf")
svm2 = SVC(kernel = "linear")
svm3 = SVC(kernel = "poly")
svm4 = SVC(kernel = "sigmoid")
print(np.mean(cross_val_score(svm1, X, data.result, cv = 10)))
print(np.mean(cross_val_score(svm2, X, data.result, cv = 10)))
print(np.mean(cross_val_score(svm3, X, data.result, cv = 10)))
print(np.mean(cross_val_score(svm4, X, data.result, cv = 10)))
```

```python
# Random Forest

rf = RandomForestClassifier()
print(np.mean(cross_val_score(rf, X, data.result, cv = 10)))
```

```python
# Decision Tree 

clf = tree.DecisionTreeClassifier()
print(np.mean(cross_val_score(clf, X, data.result, cv = 10)))
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
pca = PCA(n_components=2).fit(data1.iloc[:,0:23])
PCX=pd.DataFrame(pca.transform(data1.iloc[:,0:23]),columns=['PC1', 'PC2'])
PCX['region'] = ['j']*65+['w']*52
# np.mean(PCX['PC1'])
kmeans = KMeans(n_clusters=2, random_state=0).fit(PCX[['PC1', 'PC2']])
kmeans.labels_
```

```python
j_tn=65
w_tn=52
j_c=0
w_c=0
j_d=0
w_d=0
for i in range (len(kmeans.labels_)):
    if kmeans.labels_[i]==0:
        if PCX['region'][i]=='j':
            j_c+=1
        else:
            w_d+=1
    else:
        if PCX['region'][i]=='w':
            w_c+=1
        else:
            j_d+=1

print(max((j_c/j_tn + w_c/w_tn)/2,(j_d/j_tn + w_d/w_tn)/2))
```
